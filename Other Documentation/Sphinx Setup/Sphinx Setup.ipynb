# Instructions

1 - First things first, get Sphinx installed. 
    This is done easiest with pip install Sphinx 
    

2 - Navigate into the root directory of your project and create a new folder for 
    the resourc files Sphinx creates. It is standard to name this folder 'docs'. Navigate into this directory once 
    it is created and run 'sphinx-quickstart' (use all the default values, fill in project and author names
    accordingly)
        

This process should create the following files
        

4 - Open the conf.py file and uncomment these three lines, and make sure that the absolute path points back a directory (this is where Sphinx will look for source files). 
        

5 - Add 'modules' to the contents tree of the index.rst file so that it looks as follows. 

NOTE: The 'm' in modules needs to be DIRECTLY under the first ':' of the above ':caption:' line, and there needs to be a blank line in between. Otherwise it will throw an indentation error and will not compile correctly. 
        

6 - Now run 'sphinx-apidoc . ..'

This is a tool that will allow us to document the whole package recursively. The format for this command is "sphinx-apidoc [OPTIONS] -o [OUTPUT_PATH] [MODULE_PATH]", where for us the output path is the current directoy (.) and the module path is the parent directory (..)
       
    
(The output of this will look different when run on an actual project, and will create more files than just the modules.rst file listed above)

7 - Now all you need to do is run the command "make html" and it will build the webpage for you. You can open this easily with the command "start _build/html/index.html" 
        
    
Because this example project is empty, there is a warning about an empty toctree. You should not have this warning in your project.
